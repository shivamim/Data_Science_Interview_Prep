```python
#1  Write a program to check if a number is even or odd.

num = int(input("Enter a number:"))
if num % 2 == 0:
  print("The number is even")
else:
  print("The number is odd")
```

```python
#2 Swap two variables without a third variable.
a = 2
b = 5
print("Before swapping: a =", a, "b =", b)
a, b = b, a
print("After swapping: a =", a, "b =", b)
```

```python
#3  Check if a number is prime.
num = int(input("Enter a number:"))
if num > 1:
  for i in range(2,int(num**0.5)+1):
    if num % i == 0:
      print("The number is not prime")
      break
  else:
    print("The number is prime")
else:
  print("The number is not prime")

```

```python
#4 Factorial using recursion.
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)
num = int(input("Enter a number:"))
print("The factorial of", num, "is", factorial(num))
```

```python
# without recursion
num = int(input("Enter a number:"))
fact = 1
for i in range(1,num+1):
  fact *= i
print("The factorial of", num, "is", fact)
```

```python
#5 Generate Fibonacci sequence up to N terms.

n = int(input("Enter number of terms: "))

a, b = 0, 1

for i in range(n):
    print(a)
    a, b = b, a + b
```

```python
#6 Check if a year is a leap year.

year = int(input("Enter a year:"))
if year % 4 == 0:
  if year % 100 == 0:
    if year % 400 == 0:
      print("The year is a leap year")
    else:
      print("The year is not a leap year")
```

```python
#7 Generate fibonaci sequence

low = int(input("Enter lower limit: "))
high = int(input("Enter upper limit: "))

for n in range(low, high + 1):
    total = 0
    temp = n
    while temp > 0:
        d = temp % 10
        total = total + d*d*d
        temp = temp // 10

    if total == n:
        print(n)
```

```python
#8  Reverse an integer.
a = int(input("Enter a number:"))
rev = 0
while a > 0:
  rev = rev * 10 + a % 10
  a = a // 10
print("The reverse of the number is", rev)

```

```python
#9  Sum of digits of a number
a = int(input("Enter a number:"))
sum = 0
while a > 0:
  sum = sum + a % 10
  a = a // 10
print("The sum of the digits is", sum)
```

```python
#10 Find GCD of two numbers.

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

while b:
    a, b = b, a % b

print(a)
```

```python
#11 Temperature conversion (Menu Style)

print("Temperature Conversion Menu")
print("1. Celsius to Fahrenheit")
print("2. Fahrenheit to Celsius")
print("3. Celsius to Kelvin")
print("4. Kelvin to Celsius")

choice = int(input("Enter your choice (1-4): "))

if choice == 1:
    c = float(input("Enter Celsius: "))
    f = (c * 9/5) + 32
    print("Fahrenheit =", f)

elif choice == 2:
    f = float(input("Enter Fahrenheit: "))
    c = (f - 32) * 5/9
    print("Celsius =", c)

elif choice == 3:
    c = float(input("Enter Celsius: "))
    k = c + 273.15
    print("Kelvin =", k)

elif choice == 4:
    k = float(input("Enter Kelvin: "))
    c = k - 273.15
    print("Celsius =", c)

else:
    print("Invalid choice")

```

#Basic Python Syntax Questions- THEORETICAL

1. What is Python?


   Python is a high-level, interpreted, object-oriented programming language known for its simple syntax and readability.

2. What are the key features of Python?

   Simple syntax, interpreted, dynamically typed, object-oriented, portable, and a large standard library.

3. What is the difference between Python 2 and Python 3?

   Python 3 supports Unicode by default, better integer division, improved libraries, and is currently supported, while Python 2 is obsolete.

4. What are mutable and immutable data types?

   Mutable data types can be changed after creation (list, dictionary, set), while immutable data types cannot be changed (int, float, tuple, string).

5. Difference between list, tuple, set, and dictionary?

   List: ordered, mutable

   Tuple: ordered, immutable

   Set: unordered, unique elements

   Dictionary: key-value pairs, mutable

6. What is dynamic typing?

   Dynamic typing means the data type of a variable is determined at runtime, not at compile time.

7. What are keywords in Python?

   Keywords are reserved words with predefined meanings, such as if, else, for, while, and def.

8. Why is indentation important in Python?

   Indentation defines code blocks in Python; incorrect indentation causes syntax errors.

9. Difference between = and ==?

   = is used for assignment, while == is used for comparison.

10. What is a function?

    A function is a reusable block of code that performs a specific task and may return a value.

11. Difference between break, continue, and pass?

    break: exits the loop

    continue: skips the current iteration

    pass: does nothing (used as a placeholder)

12. What is recursion?

   Recursion is a function calling itself until a base condition is met.

13. What is a lambda function?

   A lambda function is an anonymous function written in a single expression.

14. Difference between for loop and while loop?

   A for loop is used when the number of iterations is known, while a while loop is used when the condition is unknown.

15. What are *args and **kwargs?

   *args allows passing a variable number of positional arguments, and **kwargs allows passing a variable number of keyword arguments.

16. What is an exception?

   An exception is a runtime error that interrupts the normal flow of program execution.

17. What is exception handling?

   Exception handling uses try, except, else, and finally blocks to handle runtime errors safely.

18. Difference between append() and extend()?

   append() adds a single element to a list, while extend() adds multiple elements.

19. What is the difference between shallow copy and deep copy?

   A shallow copy copies references of objects, while a deep copy copies the actual objects.
